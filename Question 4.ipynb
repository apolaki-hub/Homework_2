```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Fourier transform

```python
image = cv2.imread('beach.jpg', 0)  # Load as grayscale

#the 2D Fourier Transform
f = np.fft.fft2(image)

fshift = np.fft.fftshift(f)

magnitude_spectrum = 20 * np.log(np.abs(fshift))

plt.subplot(121), plt.imshow(image, cmap='gray')
plt.title('Input Image'), plt.xticks([]), plt.yticks([])
plt.subplot(122), plt.imshow(magnitude_spectrum, cmap='gray')
plt.title('Magnitude Spectrum'), plt.xticks([]), plt.yticks([])
plt.show()
```

### Butterworth low pass filter

```python
from numpy.fft import *
from skimage import data, exposure, img_as_float
```

```python
image = cv2.imread('beachh.jpg', 0)
cf = fftshift(fft2(image)) # fourier transform

# Butterworth low pass filter with n = 2 and D = 15.0

D = 15.0
#ar = range(-128,128)
ar = np.arange(-128,128,1.0)
x,y = np.meshgrid(ar,ar)
bl = 1.0/(1.0+((x**2+y**2)/D**2)**2)
cfbl = cf*bl
io.imshow(abs(ifft2(cfbl)))
```

### With rescaled intensity

```python
io.imshow(exposure.rescale_intensity(abs(ifft2(cfbl)),out_range=(0.0,1.0)))
```

### Butterworth High pass filter

```python
# Butterworth high pass filter with n = 2 and D = 15.0
bh = 1- 1/(1+((x**2+y**2)/D**2)**2)
cfbh = cf*bh;
io.imshow(abs(ifft2(cfbh)))
```

### With rescaled intensity

```python
io.imshow(exposure.rescale_intensity(abs(ifft2(cfbh)),out_range=(0.0,1.0)))
```

```python
from scipy import ndimage
```

### Gaussian filter on fourier transformed image with deviation of 2

```python

image = cv2.imread('beachh.jpg', 0)

# Perform Fourier Transform
fft_image = np.fft.fft2(image)

# Apply Gaussian filter in Fourier domain
sigma = 2  # Standard deviation of the Gaussian filter
filtered_fft = ndimage.fourier_gaussian(fft_image, sigma)

# Inverse Fourier Transform
filtered_image = np.fft.ifft2(filtered_fft)

# Display the results
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')


plt.subplot(1, 2, 2)
plt.imshow(np.abs(filtered_image), cmap='gray')
plt.title('Filtered Image')

plt.tight_layout()
plt.show()
```
